```python
import numpy as pd
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading data from csv file to pandas Dataframe
```

```python
mail = pd.read_csv('spam.csv', encoding='ISO-8859-1')
```

```python
print(mail)
```

```python
df = pd.DataFrame(mail)
```

```python
df
```

```python
list(df.columns)
```

```python
df = df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis =1 )
```

```python
df
```

```python
df= df.rename(columns={'v1': 'Category', 'v2': 'Message'})
```

```python
df
```

```python
mail_df = df.where((pd.notnull(df)),'')
```

```python
mail_df
```

```python
mail_df.head()
```

```python
#label spam mail as 0 and ham mail as 1
```

```python
mail_df.loc[mail_df['Category'] == 'spam', 'Category'] = 0
mail_df.loc[mail_df['Category'] == 'ham', 'Category'] = 1
```

```python
mail_df
```

```python
#separating data as texts and labels
```

```python
x = mail_df['Message']
y = mail_df['Category']
```

```python
x
```

```python
y
```

```python
#train test split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=3)
```

```python
print(x.shape)
print(x_test.shape)
print(x_train.shape)
```

```python
#FEATURE EXTRACTION
```

```python
x_train
```

```python
x_test
```

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words = 'english', lowercase =True)
```

```python
x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)
```

```python
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
# training the logistics regression model
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_features, y_train)
```

```python
#evaluating training model
```

```python
prediction = model.predict(x_train_features)
accuracy = accuracy_score(y_train, prediction)
```

```python
print('Accuracy on training data : ', accuracy)
```

```python
prediction = model.predict(x_test_features)
accuracy = accuracy_score(y_test, prediction)
```

```python
print('Accuracy on testing data : ', accuracy)
```
